```python
#ADICIONANDO ATEM E TEMPO

variaveis = ['atemp','temp']
X = bike[variaveis]
y = bike['cnt']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
model.summary()

# A primeira linha define a lista de variáveis de entrada (X) que serão usadas para ajustar um modelo de regressão linear.
# Segunda linha, a variável de saída (y) é definida como a coluna "cnt" do dataframe.
# Na terceira linha, a função "add_constant()" da biblioteca statsmodels.api é usada para adicionar uma coluna de interceptação ao conjunto de dados de entrada (X). Isso é necessário para estimar o coeficiente de interceptação do modelo de regressão linear.
# Na quarta linha, um modelo de regressão linear é criado usando a função "OLS()" da biblioteca statsmodels.api. A função recebe as variáveis de saída (y) e de entrada (X) como argumentos.
# Na quinta linha, o modelo é ajustado aos dados usando o método "fit()".
# Na última linha, é impresso um resumo estatístico do modelo usando o método "summary()" do objeto modelo. O resumo inclui informações sobre o ajuste do modelo, incluindo o coeficiente de interceptação, os coeficientes de inclinação para cada variável de entrada, os valores p associados a cada coeficiente e outras métricas de ajuste do modelo, como o coeficiente de determinação (R²) e o erro padrão da estimativa.
```
